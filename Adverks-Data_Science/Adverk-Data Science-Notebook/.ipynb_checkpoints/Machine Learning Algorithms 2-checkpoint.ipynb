# SVM Classifier

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import Perceptron, LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
from sklearn import metrics
```

## Load the Dataset

```python
bc=datasets.load_breast_cancer()
```

```python
x=bc.data
y=bc.target
```

## Create training and test split

```python
x_train, x_test, y_train, y_test =train_test_split(x, y, test_size=0.3, random_state=1, stratify=y)
```

```python
sc= StandardScaler()
sc.fit(x_train)
x_train_std=sc.transform(x_train)
x_test_std=sc.transform(x_test)
```

```python
# Instantiate the Support Vector Classifier (SVC)
svc = SVC(C=1.0, random_state=1, kernel='linear')

# Fit the model
svc.fit(x_train_std, y_train)
```

```python
y_predict=svc.predict(x_test_std)

# Measure the performance
print('Accuracy score %.3f' %metrics.accuracy_score(y_test, y_predict))
```

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import scipy.stats as stats
matplotlib.style.use('ggplot')
```

```python
# Load the mtcars data set
mtcars=pd.read_csv('/Users/rogudays/Documents/Adverks-Data_Science/Data/mtcars.csv')
mtcars.head()
```

```python
sns.heatmap(mtcars.corr(), cmap='YlGnBu', annot=True)
plt.show()
```

```python
mtcars.plot(kind="scatter",x='wt',y='mpg',figsize=(9,9),color="black")
```

```python
from sklearn import linear_model
```

```python
regression_model = linear_model.LinearRegression()
regression_model.fit(X=pd.DataFrame(mtcars['wt']), y=mtcars['mpg'])
print(regression_model.intercept_)
print(regression_model.coef_)
```

## Model Score

```python
regression_model.score(X=pd.DataFrame(mtcars['wt']), y=mtcars['mpg'])
```

## The R-squared measure

```python
train_prediction = regression_model.predict(X=pd.DataFrame(mtcars['wt']))

# Actual - prediction = Residuals
residuals=mtcars["mpg"]-train_prediction
residuals.describe()
```

```python
mtcars.plot(kind='scatter', x='wt', y='mpg', figsize=(9,9), color='black', xlim=(0,7))
# Plot regression line
plt.plot(mtcars['wt'],     # Explanitary variable
        train_prediction,  # predicted values
        color='blue');
```
