# Data Wrangling

```python
import pandas as pd

data={'Name':['Jai','Princi','Gaurav','Anuj','Ravi','Natasha','Riya'],
      'Age':[17,17,18,17,18,17,17],
      'Gender':['M','F','M','M','M','F','F'],
      'Marks':[90,76,'NaN',74,65,'NaN',71]}
df=pd.DataFrame(data)
df
```

```python
c=avg=0
for ele in df['Marks']:
    if str(ele).isnumeric():
        c+=1
        avg+=ele
avg/=c
#df.mean()
df['Marks']=df['Marks'].replace(to_replace="NaN",value=avg)
df
```

```python
df['Gender']=df['Gender'].map({'M':0,'F':1}).astype(int)
df
```

```python
df.describe()
```

```python
df=df[df['Marks']>=75]
df=df.drop(['Age'],axis=1)
df
```

## Wrangling Data Using Merge Operation

Merge Operation is used to merge raw data and into desired format

```python
import pandas as pd
details=pd.DataFrame({
    'ID':[101,102,103,104,105,106,107,108,109,110],
    'Name':['Jagroop','Praveen','Harjot','Pooja','Rahul','Nikita','Saurabh','Ayush','Dolly','Nikita',],
    'Branch':['CSE','CSE','CSE','CSE','CSE','CSE','CSE','CSE','CSE','CSE']
})
details
```

```python
fees_status=pd.DataFrame({
    'ID':[101,102,103,104,105,106,107,108,109,110],
    'Pending':[5000,250,'NIL',7383,'NIL',1010,1199,909,1,'NIL']
})
fees_status
```

```python
details=pd.merge(details,fees_status, on='ID')
details
```

```python
c=avg=0
for ele in details['Pending']:
    if str(ele).isnumeric():
        c+=1
        avg+=ele
avg/=c
#df.mean()
details['Pending']=details['Pending'].replace(to_replace="NIL",value=avg)
details
```

## Wangling Data Using Grouping Method

The grouping method in Data Analysis is used to provide results in terms 

```python
# import pandas as pd
car_selling_data={
    'Brand':['Maruti','Maruti','Maruti','Maruti','Hyundai','Hyundai','Toyota','Mahindra','Mahindra','Ford','Toyota','Ford'],
    'Year':[2010,2011,2009,2013,2010,2011,2011,2010,2013,2010,2010,2011],
    'Sold':[6,7,9,8,3,5,2,8,7,2,4,2]
}
df=pd.DataFrame(car_selling_data)
df
```

```python
# import pandas as pd
car_selling_data={
    'Brand':['Maruti','Maruti','Maruti','Maruti','Hyundai','Hyundai','Toyota','Mahindra','Mahindra','Ford','Toyota','Ford'],
    'Year':[2010,2011,2009,2013,2010,2011,2011,2010,2013,2010,2010,2011],
    'Sold':[6,7,9,8,3,5,2,8,7,2,4,2]
}
df=pd.DataFrame(car_selling_data)

grouped=df.groupby('Year')
print(grouped.get_group(2010))
```
