## Bivariate Analysis

### Gender Vs Age

```python
data1=pd.read_csv('/Users/rogudays/Documents/Adverks-Data_Science/Data/LungCap.csv')
df=pd.DataFrame(data)
```

```python
fig=plt.figure(figsize=(15,6))
sns.countplot(data=data1,x='Gender',hue='Age',palette='Accent')
plt.title('Gender vs Age')
plt.show()
```

### Age vs Caesarean, Lung Capacity vs Gender

```python
fig, axes=plt.subplots(figsize=(15,6))
sns.countplot(data=data1,x='Age',hue='Caesarean',palette='Accent')
plt.title('Age vs Caesarean')
plt.show()
```

```python
fig=plt.figure(figsize=(15,6))
sns.countplot(data=data1,x='LungCap',hue='Gender',palette='Accent')
plt.title('Lung Capacity vs Gender')
plt.show()
```

### Age vs Caesarean , Lung Capacity vs Gender

```python
fig,axes=plt.subplots(1,2,figsize=(15,6))
sns.boxplot(y='Age',data=data1,x='Caesarean',ax=axes[0])
axes[0].set_title('Caesarean vs Age',color='darkblue')
sns.boxplot(y='LungCap',data=data1,x='Gender',ax=axes[1])
axes[1].set_title('Gender vs LungCap',color='darkblue')
plt.show()
```

### Lung Capacity vs Smoke, Height vs Smoke

```python
fig,axes=plt.subplots(1,2,figsize=(15,6))
sns.violinplot(x='Smoke', data=data1, y='LungCap', ax=axes[0])
sns.violinplot(x='Smoke', data=data1, y='Height', ax=axes[1])
axes[0].set_title('Smoke vs Lung Capacity',color='darkblue')
axes[1].set_title('Smoke vs Height',color='darkblue')
plt.show()
```

### Smoke vs Age, Gender vs Height

```python
fig,axes=plt.subplots(1,2,figsize=(15,6))
sns.boxplot(x='Smoke', data=data1, y='Age', ax=axes[0])
sns.violinplot(x='Gender', data=data1, y='Height', ax=axes[1])
axes[0].set_title('Smoke vs Age',color='darkblue')
axes[1].set_title('Height vs Gender',color='darkblue')
plt.show()
```

### Lung Capacity vs Caesarean

```python
sns.swarmplot(y='LungCap',data=data1, x='Caesarean')
plt.suptitle('Lung Capacity vs Caesarean')
plt.show()
```

### Age vs Height

```python
sns.jointplot(x='Age',y='Height',data=data1,kind='reg')
plt.suptitle('Age vs height')
plt.show()
```

### Lung Capacity vs Age, Lung Capacity vs Height

```python
sns.jointplot(x='LungCap',y='Age',data=data1,kind='reg')
plt.suptitle('Lung Capacity vs Age')
plt.show()
```

```python
sns.jointplot(x='LungCap',y='Height',data=data1,kind='reg')
plt.suptitle('Lung Capacity vs Height')
plt.show()
```

### Gender vs Smoke

```python
fig=plt.figure(figsize=(8,6))
sns.countplot(x='Gender',data=data1,hue='Smoke')
plt.title('Gender vs Smoke')
plt.show()
```

```python
fig=plt.figure(figsize=(8,6))
sns.countplot(x='Smoke',data=data1,hue='Age')
plt.title('Smoke vs Age')
plt.show()
```

### Gender vs Caesarean

```python
fig=plt.figure(figsize=(10,6))
ct=pd.crosstab(index=data1['Gender'],columns=data1['Caesarean'])
plt.bar(ct.index,ct['yes'],label='Born by Caearean')
plt.bar(ct.index,ct['no'],bottom=ct['yes'],label='Not born by caaesarean')
plt.xlabel('Gender')
plt.title('Gender vs Caesarean')
plt.legend()
plt.show()
```

## Multivariate Aanalysis

```python
fig,axes=plt.subplots(2,2,figsize=(20,15))
sns.violinplot(x='Age',data=data1,y='Smoke',hue='Caesarean',ax=axes[0,0])
axes[0,0].set_title('1. Age vs Smoke')
sns.violinplot(x='Gender',data=data1,y='Age',hue='Caesarean',ax=axes[0,1])
axes[0,1].set_title('2. Gender vs Age')
sns.swarmplot(y='LungCap',data=data1,x='Age',hue='Smoke',ax=axes[1,0])
axes[1,0].set_title('3. Age vs Lung Capacity')
sns.swarmplot(y='LungCap',data=data1,x='Height',hue='Smoke',ax=axes[1,1])
axes[1,1].set_title('4. Height vs Lung Capacity')
plt.show()
```

```python
sns.lmplot(x='Age',y='LungCap',data=data1,hue='Gender',fit_reg=False,row='Smoke',col='Caesarean')
plt.show()
```
