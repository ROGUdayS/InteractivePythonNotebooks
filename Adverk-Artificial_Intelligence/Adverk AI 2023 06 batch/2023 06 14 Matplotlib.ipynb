```python
# Line Chart  : When x-axis has date/time value, y-axis is measurable value; trend analysis
# Bar Graphs  : When x-axis is categorical value, yaxis is measurable value; comparative analysis
# Scatter plot: When x-axis and y-axis are measurable values
# Pie Chart   : composition
# Histogram   : frequency
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a1=np.array([2,4,6,8,10,12])
a2=(4,7,2,8,11,3)
```

```python
plt.plot(a1,a2)
plt.show()
```

```python
plt.scatter(a1,a2)
plt.show()
```

```python
plt.bar(a1,a2)
plt.show()
```

```python
plt.barh(a1,a2)
plt.show()
```

```python
time_of_day=['Dawn','Early_morning','Morning','Noon','Afternoon','Dusk','Night','Midnight']
sales_day1=[52,43,70,30,36,42,64,28]
sales_day2=[64,28,32,56,38,24,72,46]
sales_day3=[54,62,26,48,24,50,42,24]
```

```python
plt.figure(figsize=(10,5.5))
plt.plot(time_of_day,sales_day1,color='red',label="Day1",linestyle="--",marker="*")
plt.plot(time_of_day,sales_day2,color='blue',label="Day2",linestyle="-.",marker=".")
plt.plot(time_of_day,sales_day3,color='green',label="Day3",linestyle=":",marker="o")
plt.legend(loc="upper left")
plt.xlabel("Time of the day",fontsize=11)
plt.ylabel("Sales",fontsize=11)
plt.title("Sales Trend Analysis")
plt.grid(linestyle="dashed")
plt.xticks(rotation=30)
# plt.yticks([20,25,30,35,40,45,50,55,60,65,70,75])
plt.yticks(np.arange(20,76,5))
plt.show()
```

```python
time_of_day=['Dawn','Early_morning','Morning','Noon','Afternoon','Dusk','Night','Midnight']
sales_day1=[52,43,70,30,36,42,64,28]
plt.bar(time_of_day,sales_day1,color=["#b9f542",'pink','g','lime','teal','cyan','yellow','r'],
        linestyle='dashed',edgecolor='k',width=0.8)
plt.xticks(rotation='vertical')
plt.show()
```

```python
# Colors in matplotlib
# 'r','b','g','m','y','c','k','w'
# 'red','blue','green','magenta','yellow','cyan','black','white', 'pink','lime','teal'
# #b9f542
```

```python
# Pie Chart
```

```python
pie_colors=["#b9f542",'pink','g','lime','teal','cyan','yellow','r']
exp=[0.2,0,0,0.5,0,0,0,0]
plt.pie(sales_day1,labels=time_of_day,colors=pie_colors,counterclock=False,explode=exp,shadow=True,autopct="%0.2f%%")
plt.legend(bbox_to_anchor=[1,0.5,0.5,0.5])
plt.show()
```
