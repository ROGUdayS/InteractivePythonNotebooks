```python
marks=int(input("Enter the marks scored: "))
if(marks>90):
  print("Outstanding")
  print("Good job")
elif(marks>70):
  print("Exceeds Expectations")
elif(marks>50):
  print("Met Expectations")
else:
  print("Needs Improvement")
```

```python
marks=int(input("Enter the marks scored: "))
if(marks>100):
  print("Incorrect marks. Please re-run the program and try again")
elif(marks>90):
  print("Outstanding")
  print("Good job")
elif(marks>70):
  print("Exceeds Expectations")
elif(marks>50):
  print("Met Expectations")
elif(marks>=0):
  print("Needs Improvement")
else:
  print("Incorrect marks. Please re-run the program and try again")
print("End of the program")
```

```python
import time
for i in range(0,5,1):
  print(i)
  time.sleep(1.5)
```

```python
import time as ti
for i in range(0,5,1):
  print(i)
  ti.sleep(1.5)
```

```python
from time import sleep
for i in range(0,5,1):
  print(i)
  sleep(1.5)
```

```python
from time import sleep as sl
for i in range(0,5,1):
  print(i)
  sl(1.5)
```
