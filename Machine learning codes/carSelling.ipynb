```python
fig=plt.figure(figsize=(15,6))
sns.countplot(data=data1, x='Gender',hue='Age', palette='Accent') 
plt.title("GENDER VS AGE") 
plt.show()
```

```python
fig=plt.figure(figsize=(15,6))
sns.countplot(data=data1, x='LungCap',hue='Gender', palette='Accent') 
plt.title("LungCap VS Gender") 
plt.show()
```

```python
fig,axes=plt.subplots(1,2,figsize=(15,6))
sns.boxplot(y= 'Age', data=data1, x='Caesarean', ax=axes[0])
axes[0].set_title('CAESAREAN VS AGE',color='darkblue')
sns.boxplot(y='LungCap',data=data1,x='Gender',ax=axes[1])
axes[1].set_title('GENDER VS LUNG CAPACITY', color='indigo')
plt.show()
```

```python
fig,axes=plt.subplots(1,2,figsize=(15,6))
sns.violinplot(y= 'LungCap', data=data1, x='Smoke', ax=axes[0])
axes[0].set_title('Smoke vs LungCap',color='darkblue')
sns.violinplot(y='Height',data=data1,x='Smoke',ax=axes[1])
axes[1].set_title('height vs Smoke', color='indigo')
plt.show()
```

```python
fig,axes=plt.subplots(1,2,figsize=(15,6))
sns.boxplot(y= 'Age', data=data1, x='Smoke', ax=axes[0])
axes[0].set_title('Smoke vs Age',color='darkblue')
sns.violinplot(y='Height',data=data1,x='Gender',ax=axes[1])
axes[1].set_title('Gender vs height', color='indigo')
plt.show()
```

```python
sns.swarmplot(x='Caesarean',y='LungCap',data=data1)
plt.suptitle("Lung Cap vs Caesarean")
plt.show()
```

```python
sns.jointplot(x='Age',y='Height',data=data1, kind='reg')
plt.suptitle("Age vs Height")
plt.show()
```

```python
fig=plt.figure(figsize=(10,6))
ct=pd.crosstab(index=data1['Gender'], columns=data1['Caesarean'])
plt.bar(ct.index,ct['yes'],label='Born by Caesarean')
plt.bar(ct.index,ct['no'],bottom=ct['yes'],label='Not Born by Caesarean')
plt.xlabel('Gender')
plt.title('GENDER VS CAESAREAN', color='purple')
plt.legend()
plt.show()
```
